```python
EXCEL_PATH= "/Users/amr/repos/librosa_gallery/notebooks/jasim/maqamat_songs.xlsx"
DATA_PATH= '/Users/amr/repos/librosa_gallery/notebooks/data/maqamat_v2/'
```

```python
ydl_opts = {
    'format': 'bestaudio/best',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
        'preferredquality': '192',
    }],
}




def download_youtube_mp3(URL_list, ydl_opts):
    
    with youtube_dl.YoutubeDL(ydl_opts, ) as ydl:
        ydl.download(URL_list)

# def change_dir(new_dir):
    
#     try:
#         # Change the current working Directory    
#         os.chdir(new_dir)
#         print("Directory changed")
#     except OSError:
#         print("Can't change the Current Working Directory")
```

```python
xls = pd.ExcelFile(EXCEL_PATH)
```

```python
failed_list= []

for maqam in xls.sheet_names:
    
    print('\n Downloading songs for {} maqam \n'.format(maqam))
    
    df = pd.read_excel(xls, maqam, header=None)

    
    if df.empty:
        continue
    
    URL_list= df[0].tolist()
    
    ## if URL contains a list rather than a single file, strip the list id from the URL
    URL_list= [x.split('&list=')[0] for x in URL_list]
    
    print('Downloading {} songs'.format(len(URL_list)))
    
    try: 
        
        os.chdir(os.path.join(DATA_PATH, maqam))
        download_youtube_mp3(URL_list, ydl_opts)
 
    except OSError:
        print("Can't change the Current Working Directory to {} maqam.".format(maqam))
        failed_list.append(URL_list)

    except:
        print('failed downloading songs lists for {} maqam'.format(maqam))
        failed_list.append(URL_list)
```
